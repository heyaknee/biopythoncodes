```python
#Importing Pandas to read excel file
import pandas as pd
#Importing MatPlotLib to plot a graph
import matplotlib.pyplot as plt
df = pd.read_excel("C:\\Users\\nandi\\Desktop\\Assignment\\lab2_breast_cancer.xlsx")
# Using hist function in MatPlotLib to plot a histogram
# Using .dropna() function to drop empty values in the dataset 
plt.hist(df['tumor_size'].dropna(),color="yellow",edgecolor='black')
#Label for X axis
plt.xlabel('Tumor Size')
#Label for Y axis
plt.ylabel('Frequency of Tumor Size')
#Giving title of the graph
plt.title('Tumor Size Histogram')
#Function to show the graph as output
plt.show()
```
